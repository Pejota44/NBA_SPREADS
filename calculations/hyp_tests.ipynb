```python
from statsmodels.stats.proportion import proportions_ztest
import pandas as pd
```

```python
df_playoffs = pd.read_csv("../investigation_and_cleaning/playoff_df.csv")
df_regular = pd.read_csv("../investigation_and_cleaning/RS_df.csv")
df_all = pd.read_csv("../investigation_and_cleaning/df_with_categorization.csv")
mask_regular_home = df_regular["whos_favored"] == "home"
mask_playoffs_home = df_playoffs["whos_favored"] == "home"
mask_regular_away = df_regular["whos_favored"] == "away"
mask_playoffs_away = df_playoffs["whos_favored"] == "away"
mask_all_home = df_all["whos_favored"] == "home"
mask_all_away = df_all["whos_favored"] == "away"

df_regular_home = df_regular[mask_regular_home]
df_regular_away = df_regular[mask_regular_away]
df_playoffs_home = df_playoffs[mask_playoffs_home]
df_playoffs_away = df_playoffs[mask_playoffs_away]
df_all_home = df_all[mask_all_home]
df_all_away = df_all[mask_all_away]
```

We will run 6 proportion Z-tests. Comparing:

Playoff Effect:
- All Games (RS vs PO)
- Home Favorites (RS vs PO)
- Away Favorites (RS vs PO)

Home Court Effect:
- ALL Games (Home vs Away )
- Regular Season (Home vs Away)
- Playoffs (Home vs Away)

The proportion being tested will be proportion of games with decisive outcome. Games where the favorite covers or the Underdog wins will be considered as a decisive outcome. Games where underdog covers (but loses) will be considered as not a decisive outcome. Games with an ATS result against the spread will be EXCLUDED.

```python
def run_proportion_test(df1, df2, label1, label2):

    successes1 = df1["decisive_outcome"].sum()
    successes2 = df2["decisive_outcome"].sum()

    n1 = df1["decisive_outcome"].notna().sum()
    n2 = df2["decisive_outcome"].notna().sum()

    z_stat, p_value = proportions_ztest(
        [successes1, successes2],
        [n1, n2]
    )

    pct1 = successes1 / n1
    pct2 = successes2 / n2

    print("=" * 50)
    print(f"{label1} vs {label2}")
    print(f"{label1}: {pct1:.2%}")
    print(f"{label2}: {pct2:.2%}")
    print(f"Difference: {(pct2 - pct1):.2%}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print()
```

```python
#Playoff Effect 1: Regular Season vs Playoffs (All Games)

run_proportion_test(
    df_regular,
    df_playoffs,
    "Regular Season",
    "Playoffs"
)
```

```python
#Playoff Effect 1: Regular Season vs Playoffs (Home Favorites)
run_proportion_test(
    df_regular_home,
    df_playoffs_home,
    "Home Regular",
    "Home Playoffs"
)
```

```python
#Playoff Effect 1: Regular Season vs Playoffs (Away Favorites)
run_proportion_test(
    df_regular_away,
    df_playoffs_away,
    "Away Regular",
    "Away Playoffs"
)
```

```python
#Playoff Effect 2: Home vs Away (All)
run_proportion_test(
    df_all_home,
    df_all_away,
    "Home Favorites",
    "Away Favorites"
)
```

```python
#Playoff Effect 2: Home vs Away (Regular Season)
run_proportion_test(
    df_regular_home,
    df_regular_away,
    "Home Regular",
    "Away Regular"
)
```

```python
#Playoff Effect 2: Home vs Away (Regular Season)
run_proportion_test(
    df_playoffs_home,
    df_playoffs_away,
    "Home Playoffs",
    "Away Playoffs"
)
```
